```python
import pandas as pd

# Load the dataset
file_path = 'Input/processed/dataset_reduced.csv'

df = pd.read_csv(file_path)
# Display the first few rows of the dataset to understand its structure
df.head()
```

```python
df.columns
```

```python

df['RenDys'] = (
    (df['creatinine'] > 1.3) | 
    (df['urineoutput'] < 30) |
    (df['urineoutput'] > 50) |
    (df['bun'] > 20)
).astype(int)

df['LyteImbal'] = (
    (df['calcium'] > 10.5) | 
    (df['chloride'] > 106) |
    (df['chloride'] < 98) |
    (df['potassium'] > 5.0)
).astype(int)

df['O2TxpDef'] = (
    (df['hemoglobin'] < 12)
).astype(int)

df['Coag'] = (
    (df['ptt'] > 35) | 
    (df['fibrinogen'] < 233) |
    (df['d_dimer'] > 0.5) |
    (df['thrombin'] > 20) |
     (df['inr'] > 1.5)
).astype(int)

df['MalNut'] = (
    (df['albumin'] < 3.3)
).astype(int)

df['Chole'] = (
    (df['bilirubin_direct'] > 0.3) | 
    (df['bilirubin_total'] > 1.0)
).astype(int)

df['HepatoDys'] = (
    (df['ast'] > 40) | 
    (df['alt'] > 40)
).astype(int)

df['HypGly'] = (
    (df['glucose'] > 125)
).astype(int)

df['MyoIsch'] = (
    (df['troponin_t'] > 0.04)
).astype(int)

df['CNSDys'] = (
    (df['troponin_t'] > 0.04)
).astype(int)

df['O2DiffDys'] = (
    (df['spo2'] < 92) |
    (df['fio2'] > 21)
).astype(int)

df['ThermoDys'] = (
    (df['temperature'] < 36) |
    (df['temperature'] > 38)
).astype(int)

df['Tachy'] = (
    (df['pulse'] > 90)
).astype(int)

df['VasoSprt'] = (df[['dobutamine', 'epinephrine', 'norepinephrine', 'phenylephrine', 'vasopressin']] > 0).any(axis=1).astype(int)
```
